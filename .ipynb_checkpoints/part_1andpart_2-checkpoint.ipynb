```python
# PART 1
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('data.csv', header=None)
X = data.iloc[:, 0:2].values
y = data.iloc[:, 2].values

# Initialize random weights and bias
np.random.seed(42)  # For reproducibility
w = np.random.randn(2)
b = np.random.randn()

# Learning rate
r = 0.1
print("Learning rate: ", r)
# Number of iterations
epochs = 20

# plotting function
def plot_decision_boundary(w, b, color='r', linestyle='-', alpha=1.0):
    # Create points for the line
    x1 = np.array([np.min(X[:, 0]), np.max(X[:, 0])])
    # Using w1*x + w2*y + b = 0 => y = -(w1*x + b)/w2
    x2 = -(w[0]*x1 + b)/w[1]
    plt.plot(x1, x2, color=color, linestyle=linestyle, alpha=alpha)

# Initialize the plot
plt.figure(figsize=(10, 8))

# Plot the data points
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', label='Class 1')
plt.scatter(X[y==0, 0], X[y==0, 1], color='red', label='Class 0')

# Plot initial decision boundary
plot_decision_boundary(w, b, color='red', linestyle='-', alpha=1.0)

# Training loop
for epoch in range(epochs):
    misclassified = 0
    for i in range(len(X)):
        # Calculate classification
        classification = 1 if np.dot(w, X[i]) + b > 0 else 0
        
        # Update weights and bias if misclassified
        if classification != y[i]:
            misclassified += 1
            if classification == 0:  # Should have been 1
                w = w + r * X[i]
                b = b + r
            else:  # Should have been 0
                w = w - r * X[i]
                b = b - r
                
            # Plot intermediate decision boundaries
            if epoch < epochs-1:  # Not the last epoch
                plot_decision_boundary(w, b, color='green', linestyle='--', alpha=0.3)
    
    # If no misclassifications, break
    if misclassified == 0:
        print(f"Converged at epoch {epoch}")
        break

# Plot final decision boundary
plot_decision_boundary(w, b, color='black', linestyle='-', alpha=1.0)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Perceptron Decision Boundaries\nRed: Initial, Green: Intermediate, Black: Final')
plt.legend()
plt.grid(True)
plt.show()

print("\nFinal weights:", w)
print("Final bias:", b)
```

```python
# PART 2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('data.csv', header=None)
X = data.iloc[:, 0:2].values
y = data.iloc[:, 2].values

# helper functions
def sigmoid(z):
    """Sigmoid activation function"""
    return 1 / (1 + np.exp(-z))

def predict(X, w, b):
    """Make predictions using sigmoid activation"""
    return sigmoid(np.dot(X, w) + b)

def log_loss(y_true, y_pred):
    """Compute binary cross-entropy loss"""
    epsilon = 1e-15  # Small constant to avoid log(0)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)  # Clip values to avoid log(0)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

def plot_decision_boundary(w, b, color='r', linestyle='-', alpha=1.0):
    """Plot the decision boundary"""
    x1 = np.array([np.min(X[:, 0]), np.max(X[:, 0])])
    x2 = -(w[0]*x1 + b)/w[1]  # Decision boundary where sigmoid(w1*x1 + w2*x2 + b) = 0.5
    plt.plot(x1, x2, color=color, linestyle=linestyle, alpha=alpha)

np.random.seed(42)
w = np.random.randn(2)
b = np.random.randn()

# Hyperparameters
learning_rate = 0.1
epochs = 1000
log_loss_history = []

# training with plots
# Create two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot initial data points in the first subplot
ax1.scatter(X[y==1, 0], X[y==1, 1], color='blue', label='Class 1')
ax1.scatter(X[y==0, 0], X[y==0, 1], color='red', label='Class 0')

# Plot initial decision boundary
x1 = np.array([np.min(X[:, 0]), np.max(X[:, 0])])
x2 = -(w[0]*x1 + b)/w[1]
ax1.plot(x1, x2, color='red', label='Initial boundary')

# Training loop
for epoch in range(epochs):
    # Forward pass
    y_pred = predict(X, w, b)
    
    # Compute error
    error = y - y_pred
    
    # Update weights and bias using gradient descent
    w = w + learning_rate * np.dot(X.T, error)
    b = b + learning_rate * np.sum(error)
    
    # Compute and store log loss
    current_loss = log_loss(y, y_pred)
    log_loss_history.append(current_loss)
    
    # Plot intermediate boundaries every 50 epochs
    if epoch % 50 == 0 and epoch < epochs - 1:
        x2 = -(w[0]*x1 + b)/w[1]
        ax1.plot(x1, x2, color='green', linestyle='--', alpha=0.3)

# Plot final decision boundary
x2 = -(w[0]*x1 + b)/w[1]
ax1.plot(x1, x2, color='black', label='Final boundary')

# Configure first subplot
ax1.set_xlabel('Feature 1')
ax1.set_ylabel('Feature 2')
ax1.set_title('Perceptron Decision Boundaries\nRed: Initial, Green: Intermediate, Black: Final')
ax1.legend()
ax1.grid(True)

# Plot log loss history in the second subplot
ax2.plot(range(epochs), log_loss_history, 'b-')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Log Loss')
ax2.set_title('Training Error over Epochs')
ax2.grid(True)

# Adjust layout and display
plt.tight_layout()
plt.show()

print(f"Final weights: {w}")
print(f"Final bias: {b}")
print(f"Final log loss: {log_loss_history[-1]:.6f}")

# log loss at 10-epoch intervals
plt.figure(figsize=(10, 6))
epochs_sampled = range(0, epochs, 10)
loss_sampled = log_loss_history[::10]
plt.plot(epochs_sampled, loss_sampled, 'bo-')
plt.xlabel('Epoch')
plt.ylabel('Log Loss')
plt.title('Training Error every 10 Epochs')
plt.grid(True)
plt.show()

# loss values at 10-epoch intervals
print("\nLog Loss values every 10 epochs:")
for epoch, loss in zip(epochs_sampled, loss_sampled):
    print(f"Epoch {epoch}: {loss:.6f}")
```
